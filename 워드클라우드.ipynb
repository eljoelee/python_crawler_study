```python
import nltk
```

```python
from konlpy.corpus import kolaw
from konlpy.tag import Twitter; t = Twitter()
```

```python
txt = kolaw.open('socceriline.txt').read()
```

```python
txt
```

```python
tokens_ko = t.nouns(txt)
```

```python
stop_words = ['이', '때', '그', '사람','진짜', '것', '좀','너무','다',
             '저','너무','더', '저','전','안','수','정도','제','일','거','분','내','뭐', 
            '개', '말', '또', '왜','중', '요', '고','글','보고','날','이번','함','알','듯','점','와','년','때문', '걸', '게','봐', 
             '난', '곳', '놈', '임', '후', '쪽', '대한', '명', '시', '도', '데', '국','가장', '원','이건','줄', '위','요즘',
              '이제','물론','너','나','추','정말','애','재','보','아직','월', '기준', '위해','거의', '세', '전부', '달', '등', '은', '살', '한번'
             '그거', '둘', '건', '만', '씨', '뭔가', '누가', '대해', '예전', '네', '갑자기', '주', '거기', '부', '어제', '클릭', '의', '며',
             '지난','감','여','모해','기','번','꼭','를','앞','이후','천','여기','기','볼','에서','음','그게','라며','스','해도','무슨','두',
             '바','초','진','억','친','뿐','건가','그거','그냥','못','선','나름','일단','어디',]
tokens_ko = [each_word for each_word in tokens_ko if each_word not in stop_words]
```

```python
tokens_ko
```

```python
ko = nltk.Text(tokens_ko, name="싸줄")
ko.vocab().most_common(200)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import pytagcloud
```

```python
tag = ko.vocab().most_common(200)
```

```python
taglist = pytagcloud.make_tags(tag, maxsize=80)
```

```python
taglist
```

```python
pytagcloud.create_tag_image(taglist, 'wordcloud.jpg', size=(900, 600), fontname='Korean', rectangular=False)
```
